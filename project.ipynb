#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# TODO

# Function to normalize prices
def normalize_prices(data):
    return (data - data.min()) / (data.max() - data.min())

# Apply normalization to the price within each population
south_belt_houses['normalized_price'] = south_belt_houses.groupby('level5')['price'].transform(normalize_prices)

# Plotting histograms for each population
populations = south_belt_houses['level5'].unique()
for pop in populations:
    plt.hist(south_belt_houses[south_belt_houses['level5'] == pop]['normalized_price'], bins=20, alpha=0.5, label=pop)

plt.title('Normalized Price Distributions by Population')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

The plot shows the normalized price distributions for four populations of south belt houses. Each population's distribution is represented by a histogram on the same graph, allowing for direct comparison. When comparing the populations, one can infer that Getafe's housing market is potentially more varied, with a range of affordable to more expensive houses. Fuenlabrada and Alcorcón seem to be more dominated by lower-priced houses, while Leganés presents a more balanced but less concentrated price range.

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆)

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

# List of populations
populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Create a subplot for each population
fig, axs = plt.subplots(2, 2, figsize=(10, 10)) 
axs = axs.flatten() 

for i, pop in enumerate(populations):
    # Filter the subset for the current population
    subset = south_belt_houses[south_belt_houses['level5'] == pop]
    
    # Scatter plot for current population
    axs[i].scatter(subset['surface'], subset['price'], alpha=0.5)
    
    # Set title and labels
    axs[i].set_title(f'Price vs. Surface Area in {pop}')
    axs[i].set_xlabel('Surface Area')
    axs[i].set_ylabel('Price')


plt.tight_layout()
plt.show()
```
